### Worklist
> 1. Get dummy variables from movie (get domain)
> 2. Split train, test and valid dataset
> 3. Further Split three parts into 20 domains.

### Part One

```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
pd.options.mode.chained_assignment = None 
```

```python
df=pd.read_csv('movies.csv')
df_new=df['genres'].str.get_dummies()
df=pd.concat([df,df_new],axis=1)
df.to_csv('movie_type_divide.csv')
```

### Part two

```python
train_index = []
with open('train_index.txt', 'r') as f:
    for line in f:
        train_index.extend(line.split(','))

valid_index = []
with open('valid_index.txt', 'r') as f:
    for line in f:
        valid_index.extend(line.split(','))
        
test_index = []
with open('test_index.txt', 'r') as f:
    for line in f:
        test_index.extend(line.split(','))

data=pd.read_csv('ratings.csv')
data=data.drop('timestamp',1)

train_index=[int(i[:-1]) if i[-1:]=='\n' else int(i) for i in train_index]
valid_index=[int(i[:-1]) if i[-1:]=='\n' else int(i) for i in valid_index]
test_index=[int(i[:-1]) if i[-1:]=='\n' else int(i) for i in test_index]

data_train=data.iloc[train_index]
data_valid=data.iloc[valid_index]
data_test=data.iloc[test_index]

data_train.to_csv('data_train.csv')
data_valid.to_csv('data_valid.csv')
data_test.to_csv('data_test.csv')
```
